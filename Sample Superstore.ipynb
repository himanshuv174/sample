```python
# Unique values for country column
pd.DataFrame(df.Country.unique(), columns=["Countries"])
```

```python
# Unique values for State column
pd.DataFrame(df.State.unique(), columns=["States"]).count()
```

```python
region_sales = df.groupby("Region").sum()[["Sales"]]
region_sales
```
